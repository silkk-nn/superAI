# Tokenize the data

```
from pythainlp.tokenize import word_tokenize
from tqdm import tqdm
```

```
%%time
tokenized_lines = []

with open('thwiki-sentseg-small.txt', 'r') as f:
    lines = f.readlines()
    for line in tqdm(lines):
        if line == '\n':
            continue
        else:
            tokens = word_tokenize(line)
            tokenized_lines.append(tokens)
```

```
with open('thwiki-tokenized-small.txt', 'w') as f:
    for token_line in tokenized_lines:
        joint_line = ''
        for token in token_line:
            joint_line += (token + ' ')
        f.write(joint_line)
```
